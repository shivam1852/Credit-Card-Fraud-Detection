```python
#Importing Dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

```python
#Load The Credit_data.csv file
credit_card_data = pd.read_csv('C:/Users/lenovo/Desktop/CodeClause/creditcard.csv')

#check first 5 rows of the data set
credit_card_data.head()
```

```python
# Check number of missing value in each coloum
credit_card_data.isnull().sum()
```

```python
# check distribution of legit transcation and fraudlent transcation
credit_card_data['Class'].value_counts()
```

```python
#seprating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statstical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for Both transcation
credit_card_data.groupby('Class').mean()
```

```python
#build sample dataset 
legit_sample = legit.sample(n=492)
```

```python
#concatenating two dataframes
new_dataset = pd.concat([legit_sample,fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Splitting the data into features and targets

X = new_dataset.drop(columns='Class',axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split Data into train and Testing data

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, stratify = Y,random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

```python
#Model Training

#Logistic Regression

model = LogisticRegression(max_iter=10000)  # Increase the number of iterations
```

```python
#Training the logistic regression model with training data
model.fit(X_train, Y_train)
```

```python
#Model Evaluation
#Accuracy Score

#Accuracy on Training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy On Training Data : ',training_data_accuracy)
```

```python
#Accuracy on Test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy On Test Data : ',test_data_accuracy)
```
